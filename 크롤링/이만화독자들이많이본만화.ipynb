```python
import pandas as pd
```

```python
# 필요 모듈 임포트
from urllib.request import Request, urlopen
from bs4 import BeautifulSoup
import numpy as np
```

```python
link = pd.read_csv('image_table.csv')
```

```python
link2 =pd.read_csv('old_image_table.csv')
```

```python
origin = pd.concat([link, link2])
```

```python
links = origin.page_link
```

```python
url
```

```python
soup.find_all(class_='asideBox')
```

```python
# 이미지 정보 테이블 작품명 / 이미지(각화 작품 썸네일 주소) / 제목(~화) / 각화 주소
mostview_table = origin.copy()
recommend = []
for link in links:
    url = link
    request = Request(url)
    request.get_method = lambda: 'GET'
    response_body = urlopen(request).read()
    soup = BeautifulSoup(response_body, 'html.parser')
    if soup.find_all(class_='asideBox') == '[]':
        recommend.append('[]')
    else :
        recommend.append(soup.find_all(class_='asideBox')[-1])

mostview_table['recommend'] = recommend
```

```python
print(mostview_table.recommend.values[0])
```

```python
mostview = mostview_table.reset_index(drop = True)
```
